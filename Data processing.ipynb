```python
# Function to calculate missing values by column
def missing_values_check(df):
        # Total missing values
        missing = df.isnull().sum()
        
        # Percentage of missing values
        missing_percent = 100 * df.isnull().sum() / len(df)
        
        # Make a table with the results
        missing_output = pd.concat([missing, missing_percent], axis=1)
        
        # Rename the columns
        missing_output= missing_output.rename(columns = {0 : 'Missing Values', 1 : '% of Total Values'})
        
        # Sort the table by percentage of missing descending
        missing_output = missing_output[ missing_output.iloc[:,1] != 0].sort_values('% of Total Values', ascending=False).round(1)
        
        # Print some summary information
        print ("Your selected dataframe has " + str(df.shape[1]) + " columns.\n""There are " + str(missing_output.shape[0]) +" columns that have missing values.")
        
        # Return the dataframe with missing information
        return missing_output
```

```python
# Fill NaN with "Unknown"
Train["composer"].fillna("Unknown",inplace=True)
Test["composer"].fillna("Unknown",inplace=True)
```

```python
# Change "Composer Unknown" into "Unknown"
Train["composer"].replace("Composer Unknown","Unknown",inplace=True)
Test["composer"].replace("Composer Unknown","Unknown",inplace=True)
```

```python
print(Train["source_screen_name"].unique())
print(Test["source_screen_name"].unique())
print(set(list(Train["source_screen_name"].unique())) ^ set(list(Test["source_screen_name"].unique())))
```

```python
# Fill NaN with "Unknown"
Train["source_screen_name"].fillna("Unknown",inplace=True)
Test["source_screen_name"].fillna("Unknown",inplace=True)
```

```python
# Change "people local or global" into "Unknown"
Train["source_screen_name"].replace("People local","Unknown",inplace=True)
Test["source_screen_name"].replace("People global","Unknown",inplace=True)
```

```python
# Also fill NaN with Unknown or 0 for Source_type/Source System Tab/Language/Artist Name
Train["source_type"].fillna("Unknown",inplace=True)
Test["source_type"].fillna("Unknown",inplace=True)
Train["source_system_tab"].fillna("Unknown",inplace=True)
Test["source_system_tab"].fillna("Unknown",inplace=True)
Train["language"].fillna(0,inplace=True)
Test["language"].fillna(0,inplace=True)
Train["artist_name"].fillna("Unknown",inplace=True)
Test["artist_name"].fillna("Unknown",inplace=True)
```

```python
# Use test mean value to fill the NaN for song_length
mean_song_length = Train["song_length"].mean()
Train["song_length"].fillna(mean_song_length,inplace=True)
Test["song_length"].fillna(mean_song_length,inplace=True)
```

```python
#Gener has "|" cell, we take first on style if existing "|".Others fill NaN with 0
Train["genre_ids"].fillna(0,inplace=True)
Test["genre_ids"].fillna(0,inplace=True)
def split_genre(ids):
    temp = ids.split("|")
    return(temp[0])
Train["genre_ids"] = Train["genre_ids"].apply(lambda ids : split_genre(str(ids)))
Test["genre_ids"] = Test["genre_ids"].apply(lambda ids : split_genre(str(ids)))
```

```python
#Delete Lyricist because of the over 50% missing rate.
Traindata = Train[["source_system_tab","source_screen_name","source_type","song_length", "genre_ids","artist_name","composer", "language", "city", "age", "gender", "registered_via", "target"]]
Testdata = Test[["source_system_tab","source_screen_name","source_type","song_length", "genre_ids","artist_name","composer", "language", "city", "age", "gender", "registered_via"]]
```
